*Find Which Variables are Most Correlated With Each Pollutant*

```python
target_pollutants = ['CO(GT)', 'C6H6(GT)', 'NOx(GT)', 'NO2(GT)']

for target in target_pollutants:
    correlations = []
    for col in df.columns:
        if col != target:
            corr, p_value = pearsonr(df[target], df[col])
            correlations.append((col, corr, p_value))

    # Sort by absolute correlation value
    correlations.sort(key=lambda x: abs(x[1]), reverse=True)

    print(f"\nTop correlations with {target}:")
    for col, corr, p_value in correlations[:5]:
        print(f"{col}: r = {corr:.4f}, p = {p_value:.4e}")

# Print key findings
print("\nKey EDA Findings:")
print("1. Distributions of pollutants")
print("2. Temporal patterns in air quality")
print("3. Correlations between pollutants and weather variables")
print("4. Hourly and weekly patterns in pollutant concentrations")
```

*Model Implementation and Evaluation*

take one target pollutant

```python
target = 'C6H6(GT)'  # Benzene concentration
```

Define target variables to predict

```python
target_pollutants = ['CO(GT)', 'C6H6(GT)', 'NOx(GT)', 'NO2(GT)']
```

Feature Selection : sensor readings, weather variables, time features

```python
features = [col for col in df_features.columns if col not in target_pollutants]
```

Prepare X, y

```python
X = df_features[features]
y = df_features[target]
```

Split data for training and testing (using time-based split)

```python
# We'll use the first 70% of the data for training, the remaining 30% for testing
train_size = int(0.7 * len(df_features))
X_train, X_test = X.iloc[:train_size], X.iloc[train_size:]
y_train, y_test = y.iloc[:train_size], y.iloc[train_size:]

print(f"Training data size: {X_train.shape}")
print(f"Testing data size: {X_test.shape}")
```

Feature Scaling

```python
scaler = StandardScaler()
X_train_sc = scaler.fit_transform(X_train)
X_test_sc = scaler.transform(X_test)
```

*Model Initialization*

```python
models = {
    'Linear Regression': LinearRegression(),
    'Gradient Boosting Regressor': GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, random_state=42),
    'Random Forest Regressor': RandomForestRegressor(n_estimators=100, random_state=42)
}
```

Train and evaluate each model

```python
results = {}

for name, model in models.items():
    # Train the model
    print(f"\nTraining {name}...")
    model.fit(X_train_sc, y_train)

    # Make predictions
    y_pred = model.predict(X_test_sc)

    # Evaluate the model
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    results[name] = {
        'MSE': mse,
        'RMSE': rmse,
        'MAE': mae,
        'R': r2
    }

    print(f"{name} Results:")
    print(f"  MSE: {mse:.4f}")
    print(f"  RMSE: {rmse:.4f}")
    print(f"  MAE: {mae:.4f}")
    print(f"  R: {r2:.4f}")
```

Plot actual vs predicted values for each model

```python
for name, model in models.items():
    y_pred = model.predict(X_test_sc)

    plt.figure(figsize=(10, 6))
    plt.plot(y_test.index, y_test.values, label='Actual', color='blue')
    plt.plot(y_test.index, y_pred, label='Predicted', color='red', alpha=0.7)
    plt.title(f'{name} - Actual vs Predicted {target}')
    plt.xlabel('Date')
    plt.ylabel(target)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

Feature importance for Random Forest Regressor

```python
rf_model = models['Random Forest Regressor']
feature_importance = pd.DataFrame({
    'Feature': features,
    'Importance': rf_model.feature_importances_
})
feature_importance = feature_importance.sort_values('Importance', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=feature_importance.head(15))
plt.title('Random Forest - Top 15 Feature Importances')
plt.tight_layout()
plt.show()
```

Feature Importance for Gradient Boosting Regressor

```python
rf_model = models['Gradient Boosting Regressor']
feature_importance = pd.DataFrame({
    'Feature': features,
    'Importance': rf_model.feature_importances_
})
feature_importance = feature_importance.sort_values('Importance', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=feature_importance.head(15))
plt.title('Gradient Boosting - Top 15 Feature Importances')
plt.tight_layout()
plt.show()
```

*Compare Model Performances*

```python
# Compare model performances
metrics = ['MSE', 'RMSE', 'MAE', 'R']
model_comparison = pd.DataFrame(index=models.keys(), columns=metrics)

for model_name, metrics_dict in results.items():
    for metric, value in metrics_dict.items():
        model_comparison.loc[model_name, metric] = value

print("\nModel Comparison:")
print(model_comparison)

# Create a bar chart to compare model performance
plt.figure(figsize=(15, 10))
for i, metric in enumerate(metrics, 1):
    plt.subplot(2, 2, i)
    ax = sns.barplot(x=model_comparison.index, y=model_comparison[metric])
    plt.title(f'Model Comparison - {metric}')
    plt.xticks(rotation=45)
    plt.ylabel(metric)

plt.tight_layout()
plt.show()
```

**Analysis**

The exceptionally high R values (0.999999 for Random Forest and 0.999996 for Gradient Boosting) suggest that these models have almost perfectly captured the patterns in the test data. However, such near-perfect scores warrant further investigation:

1 Potential Overfitting
The Random Forest and Gradient Boosting models may be overfitting, capturing noise rather than just meaningful patterns.

Tree-based models, especially those with high depth, can memorize training data instead of generalizing well.

The extremely low MSE, RMSE, and MAE values indicate that the models might be too optimized for this dataset.

2 Feature-Target Correlation
A strong correlation between the features and the target variable could explain the near-perfect accuracy.

The dataset might contain highly predictive features, making it easier for the models to achieve low error rates.

Further feature importance analysis is needed to determine which attributes are driving the predictions.

3 Data Splitting Considerations
If a time-based train-test split was used, similar patterns might exist in both sets, leading to artificially high accuracy.

A random split or cross-validation should be used to verify the true generalization performance of these models.

4 Linear Regression's Higher Errors
The Linear Regression model performed significantly worse, with:

MSE: 15.925267

RMSE: 3.990647

MAE: 3.632564

R Score: 0.994945

This suggests that the relationship between the features and target variable is non-linear, which tree-based models capture better.

**Conclusion**

Random Forest showed the best performance with the lowest error values.  
Gradient Boosting was also highly effective, though slightly behind Random Forest.  
Linear Regression had the highest errors, making it less suitable for this dataset.
